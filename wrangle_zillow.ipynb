```python
df = df0.replace(r'^\s*$', np.nan, regex=True)
```

```python
df = df.dropna()
print(df.info())
print("----------------------------")
print(df.isnull().sum(), "\n \t nulls")
print("----------------------------")
df.describe().T
```

```python
##ranges on these look very crazy, diving deeper
for col in df.columns:
    print(col, "{:,}".format((df[col].describe()["max"]- df[col].describe()["min"]).astype(int)))
```

```python
## def not looking correct. a single family house with 25 bedrooms? 32 bathrooms? almost a million sqft? 90mil in taxes a year?
## if looking for model to predict, i'm going to remove outliers that would skew data
## using 3x STD from MEAN to determine retention

## dropping 0 bathroom counts, since that's a code violation for a single family unit (requirement)
## also dropping 0 bedroom counts, since studios should not be considered a single family residence

## dropping any homes with less than 320sqft due to International Residential Code (IRC). 
## The IRC codes require that all homes must be built on a minimum of 320 square feet.

## using 3 signma for deviation allowance to capture 99.7% (removing the extreme outliers)

remove_list = ['bedroomcnt', 'bathroomcnt', 'calculatedfinishedsquarefeet',
       'taxvaluedollarcnt', 'yearbuilt', 'taxamount']

for col in remove_list:
    df = df[(df[col] < (df[col].describe()["std"]*3 + df[col].describe()["mean"])) & (df[col] > (df[col].describe()["mean"] - df[col].describe()["std"]*3))]
df = df[df.bathroomcnt>0]
df = df[df.bedroomcnt>0]
df = df[df.calculatedfinishedsquarefeet>320]

df.describe().T
```

```python
## looks much better, only lost a few thousand records, still unsure why some of the values are so high
for col in df.columns:
    print(col, "{:,}".format((df[col].describe()["max"]- df[col].describe()["min"]).astype(int)))
df.shape
```

```python
##checking the distribution now that the outliers are removed

plt.figure(figsize=(25, 3))

# List of columns
cols = df.columns.tolist()

for i, col in enumerate(cols):

    # i starts at 0, but plot nos should start at 1
    plot_number = i + 1 

    # Create subplot.
    plt.subplot(1,len(cols), plot_number)

    # Title with column name.
    plt.title(col)

    # Display histogram for column.
    df[col].hist(bins=5)

    # Hide gridlines.
    plt.grid(False)
```

```python
bin_df = pd.DataFrame()
temp_list = ['bedroomcnt', 'bathroomcnt',"taxvaluedollarcnt", 'calculatedfinishedsquarefeet',
        'yearbuilt', 'taxamount']
for col in temp_list:
    bin_df[f"{col} binned"] = pd.cut(df[col],5)
bin_df["fips"] = df.fips
bin_df["dummy"] = 1
bin_df
```

```python
binned = bin_df.groupby(['bedroomcnt binned', 'bathroomcnt binned', 'taxvaluedollarcnt binned',
       'calculatedfinishedsquarefeet binned', 'yearbuilt binned',
       'taxamount binned', 'fips']).sum().reset_index()
binned = binned[binned.dummy > 0]
```

```python
import plotly.express as px
import plotly

s_list = [  'fips',
            'yearbuilt binned', 
            'calculatedfinishedsquarefeet binned',
            'bedroomcnt binned', 
            'bathroomcnt binned',  
            'taxvaluedollarcnt binned',
            
        ]

all = binned["dummy"].sum()
binned["percentage"] = (round(((binned["dummy"]/all)*100),3).astype(str) + "%")



fig = px.sunburst(  binned, 
                    path=s_list,
                    labels=s_list,
                    values='dummy',
                    color = "dummy"
                    )
plotly.offline.iplot(fig, validate=False)
```
