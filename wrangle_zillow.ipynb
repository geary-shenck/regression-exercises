```python
df = df0.replace(r'^\s*$', np.nan, regex=True)
```

```python
df = df.dropna()
print(df.info())
print("----------------------------")
print(df.isnull().sum(), "\n \t nulls")
print("----------------------------")
df.describe().T
```

```python
##ranges on these look very crazy, diving deeper
for col in df.columns:
    print(col, "{:,}".format((df[col].describe()["max"]- df[col].describe()["min"]).astype(int)))
```

```python
## def not looking correct. a single family house with 25 bedrooms? 32 bathrooms? almost a million sqft? 90mil in taxes a year?
## if looking for model to predict, i'm going to remove outliers that would skew data
## using 3x STD from MEAN to determine retention

## dropping 0 bathroom counts, since that's a code violation for a single family unit (requirement)
## also dropping 0 bedroom counts, since studios should not be considered a single family residence

## dropping any homes with less than 320sqft due to International Residential Code (IRC). 
## The IRC codes require that all homes must be built on a minimum of 320 square feet.

for col in df.columns:
    df = df[(df[col] < (df[col].describe()["std"]*6 + df[col].describe()["mean"])) & (df[col] > (df[col].describe()["mean"] - df[col].describe()["std"]*6))]
df = df[df.bathroomcnt>0]
df = df[df.bedroomcnt>0]
df = df[df.calculatedfinishedsquarefeet>320]

df.describe().T
```
